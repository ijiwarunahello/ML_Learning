```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
# 腫瘍の悪性・良性を判断するbreast_cancer
breast_cancer = load_breast_cancer()
X = breast_cancer.data
y = breast_cancer.target
```

```python
# ロジスティック識別
clf1 = LogisticRegression()
clf1.fit(X, y)
```

```python
# Warningは一旦無視
```

```python
for f, w in zip(breast_cancer.feature_names, clf1.coef_[0]):
    print("{0:<23}: {1:6.2f}".format(f, w))
```

```python
# 決定木
clf2 = DecisionTreeClassifier(max_depth=2)
clf2.fit(X, y)
```

```python
# 決定木を可視化
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(15, 10))
plot_tree(clf2, feature_names=breast_cancer.feature_names, filled=True)
plt.show()
```

腫瘍の半径の平均（radius）が16.795より小さいか大きいかで分けている

小さい場合、凹みの数の平均（concave points）を見て、0.136より小さければ良性、大きければ悪性と判断

一方、腫瘍半径の平均が16.795より大きい場合は、濃淡値の標準偏差（texture）を見て、16.11より小さければ良性、大きければ悪性と判断
